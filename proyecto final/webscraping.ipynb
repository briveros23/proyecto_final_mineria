```python
match_id_df = pd.read_csv('MatchId.csv')
matchid_list = match_id_df["MatchId"]
performance_list = []

def get_match_json(matchid):
  url_pull_match = "https://{}.api.riotgames.com/lol/match/v5/matches/{}?api_key={}".format(Continent,matchid,api_key)
  match_data_all = requests.get(url_pull_match).json()
  match_type = match_data_all["info"]["gameMode"]
  if match_type == "CLASSIC":
    for participant in range(0,10):
      time.sleep(1.1)
      player=match_data_all["info"]["participants"][participant]
      sum_id_participants = player["summonerId"]
      url_sum = "https://{}.api.riotgames.com/lol/league/v4/entries/by-summoner/{}?api_key={}".format(region,sum_id_participants,api_key)
      sum_pull = requests.get(url_sum).json()
      global elo
      if sum_pull == []:
        elo = "unrancked"
      elif sum_pull[0]["queueType"] == "RANKED_FLEX_SR":
        if len(sum_pull) == 1:
          elo = "unrancked"
        else:
          elo = "{} {}".format(sum_pull[1]["tier"],sum_pull[1]["rank"])
      elif sum_pull[0]["queueType"] == "RANKED_SOLO_5x5":
        elo = "{} {}".format(sum_pull[0]["tier"],sum_pull[0]["rank"]) 

      player_info = elo, player["championName"], player["win"], player["kills"], player["assists"], player["deaths"], player["individualPosition"], player["totalMinionsKilled"], player["goldEarned"], player["dragonKills"], player["totalDamageDealtToChampions"], player["totalDamageTaken"], player["turretTakedowns"], player["baronKills"], player["damageDealtToBuildings"], player["damageDealtToObjectives"],player["detectorWardsPlaced"], player["inhibitorTakedowns"], player["neutralMinionsKilled"], player["wardsKilled"] ,player["wardsPlaced"], player["visionScore"], player["summoner2Id"], player["summoner1Id"]
      performance_list.append(player_info)
      
i=1
for match in matchid_list:
  time.sleep(1.5)
  if i%(4980/10) == 0:
    print("x",end =" ")
  i+=1   
  get_match_json(match)

```
